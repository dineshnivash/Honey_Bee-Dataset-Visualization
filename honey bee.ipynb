```python
# 3) Display the percentage distribution of the data in each year using the pie chart.
labels=df['year'].value_counts()*100
plt.pie(labels,autopct='%.1f%%',labels=labels.index);
```

```python
# 4) Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.
sns.displot(df['priceperlb'],kind='hist')
plt.title('Distribution of Price per lb')
plt.show()
```

```python
#The majority of the prices fall between 2 and 4 dollars per pound.
#This information can be useful for understanding the pricing trends in the honeybee production industry.
```

```python
# 5) Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.
df[['numcol','prodvalue']].plot(x='numcol',y='prodvalue',title='relation between  numcol and prodvalue',kind='scatter',figsize=[12,5])
plt.show()
```

```python
#it has linear relationship
#In this the numcol increases the prodvalue also increases. There is positive relation between the numcol and prodvalue in a honeybee production
```

```python
# 6) Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.
df[['prodvalue','year']].plot(x='year',y='prodvalue',title='Relation between year and prodvalue',kind='box',figsize=[12,5])
plt.show()
#it has some of outliers in a data
#there is a positive relationship between the year and honeybee production values.
```

```python
# 7) Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')
sns.pairplot(df[['numcol','yieldpercol','totalprod','prodvalue','year']],diag_kind='kde',corner=True) #diag_kind is used to get the plot what we want in diagonals
#It can help us identify patterns and trends in the data
#there is a positive relationship between 'yieldpercol' and 'prodvalue'
```

```python
# 8) Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue')
df_corr=df[['numcol', 'yieldpercol','totalprod', 'stocks', 'priceperlb', 'prodvalue']]
sns.heatmap(df_corr.corr(),annot=True)
plt.show()
```

```python
#There is a strong positive correlation between 'totalprod' and 'numcol'
#There is a negative correlation between 'priceperlb' and 'totalprod'
```
